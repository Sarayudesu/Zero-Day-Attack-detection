```python
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8])
combined_df.head()
```

```python
shuffled_df = combined_df.sample(frac=1, random_state=42)
shuffled_df = shuffled_df.reset_index(drop=True)
```

```python
shuffled_df.info()
```

```python
dataset=shuffled_df.dropna()
dataset.head()
```

```python
labeldf=dataset[' Label']
#change the label column
newlabeldf=labeldf.replace({ 'BENIGN' : 0, 'DoS Hulk' : 1,'PortScan' : 2,'DDoS' : 3,'DoS GoldenEye' : 4,
        'FTP-Patator' : 5,'SSH-Patator' : 6,'DoS slowloris' : 7,'DoS Slowhttptest' : 8,'Bot' : 9, 'Web Attack � Brute Force' : 10,
        'Web Attack � XSS' : 11, 'Infiltration' : 12, 'Web Attack � Sql Injection' : 13, 'Heartbleed' : 14})
```

```python
dataset[' Label'] = newlabeldf
```

```python
df=dataset.drop_duplicates()
```

```python
x=df.drop(' Label',axis=1)
y=df[' Label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train[np.isinf(x_train)] = np.nan
x_train[np.abs(x_train) > np.finfo(np.float64).max] = np.nan

x_test[np.isinf(x_test)] = np.nan
x_test[np.abs(x_test) > np.finfo(np.float64).max] = np.nan

# Drop rows with NaN values
x_train = x_train[~np.isnan(x_train).any(axis=1)]
x_test = x_test[~np.isnan(x_test).any(axis=1)]

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
```

```python
print("Number of samples in x_train_scaled:", x_train_scaled.shape[0])
print("Number of samples in y_train:", y_train.shape[0])
```

```python
y_train_selected = y_train[:x_train_scaled.shape[0]]
```

```python
mlp.fit(x_train_scaled, y_train_selected)
```

```python
y_pred = mlp.predict(x_test_scaled)
```

```python
print("Number of samples in y_test:", len(y_test))
print("Number of samples in y_pred:", len(y_pred))

```

```python
y_test_selected= y_train[:y_pred.shape[0]]
```

```python
accuracy = accuracy_score(y_test_selected, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test_selected, y_pred, average='weighted')
print("Precision:", precision)
recall = recall_score(y_test_selected, y_pred, average='weighted')
print("Recall:", recall)
f1 = f1_score(y_test_selected, y_pred, average='weighted')
print("F1-score:", f1)
```
